# Preprocessing for OQ

```python
import pandas as pd
pd.low_memory = False
```

### Load and clean scraped data

```python
df_primo = pd.read_csv("../data_collection/data/prima-parte.csv", na_filter=True, na_values='[]')
df_secondo = pd.read_csv("../data_collection/data/SecondaParte.csv", na_filter=True, na_values='[]')
df_terzo = pd.read_csv("../data_collection/data/terza-parte.csv", na_filter=True, na_values='[]')
df_quarto = pd.read_csv("../data_collection/data/quarta_parte.csv", na_filter=True, na_values='[]')
```

```python
joinDF = [df_primo, df_secondo, df_terzo, df_quarto]
```

```python
df = pd.concat(joinDF)
del df["Unnamed: 0"]
df = df.drop(["id", "user_id","retweet_date", "translate", "conversation_id", "trans_src", "trans_dest","link", "near", "geo", "source", "user_rt_id","user_rt", "retweet_id", "place", "thumbnail", "video", "language","reply_to", "photos" , "created_at", "timezone", "cashtags", "urls", "retweet", "quote_url"], axis=1)
```

### Load and process community DF

```python
df_community = pd.read_csv("../data_collection/data/Community_louvain.csv", na_filter=True, na_values='[]')
```

```python
df_community
```

```python
dict_comm = df_community.to_dict(orient="list")
```

```python
dict_comm_transformed = {user: community for community, users in dict_comm.items() for user in users}
```

```python
df_community = pd.DataFrame.from_dict(dict_comm_transformed, orient='index',columns=["community"])
```

```python
df_community = df_community.rename_axis('username').reset_index()
```

```python
dict_comm_transformed
```

```python
df_community['original_community_extension'] = df_community['community'].map(df_community['community'].value_counts())
```

```python
df_community
```

### Merge on username

```python
df_nlp = (pd.merge(df, df_community, left_on='username', right_on='username', how='inner').drop('username', axis=1))
```
